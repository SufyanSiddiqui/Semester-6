# Subplots

The command
plt.subplot () needs three arguments to be passed into it:
The number of rows of subplots
•
The number of columns of subplots
•
The index of the subplot we want to create
For instance, the command
plt.subplot (2, 2 ) would create “ Subplot 2 ” from
the figure above

```python
# Data sets
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
plt.figure(figsize=(10,8))
# First Subplot
plt.subplot(2, 2,1)
plt.plot(x, y, color='green')
plt.title('First')
# Second Subplot
plt.subplot(2, 2,2)
plt.plot(x, y, color='steelblue') 
plt.title('Second')

plt.subplot(2, 2,3)
plt.plot(x, y, color='red') 
plt.title('Third')

plt.subplot(2, 2,4)
plt.plot(x, y, color='cyan') 
plt.title('Fourth')
# Display both subplots
plt.show()
```

```python
months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
plt.plot(months,temperature,color='red',marker='o')
plt.subplots_adjust(wspace = 0.5)
plt.subplot(2,2,3)
plt.plot(temperature,flights_to_hawaii,color='purple',marker='o')
plt.subplot(2,2,4)
plt.plot(months,temperature,color='cyan',marker='x')

plt.show()

```

# Legends and Ticks

```python
ax = plt.subplot()
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9,16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27,64])
plt.legend(['Parabola','Cubic'])
ax.set_xticks([1, 2,4])
```

```python
ax = plt.subplot()
plt.plot([1, 3, 3.5], [0.1, 0.6, 0.8],marker='s')
ax.set_yticks([0.1, 0.6, 0.8])
ax.set_yticklabels(['10%', '60%', '80%'])
plt.show()
```

# Bar Chart

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales = [91, 76, 56, 66, 52, 27]
plt.figure(figsize=(15,8))
ax=plt.subplot()
plt.bar( range(len(drinks)),sales,color='green')
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks,rotation=90)
```
