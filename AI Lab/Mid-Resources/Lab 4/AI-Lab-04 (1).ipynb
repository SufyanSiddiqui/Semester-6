```python

def Triangles(graph):
    vertices = len(graph)
    count = 0
     
   
    for i in range(vertices):
        for j in range(vertices):
            for k in range(vertices):
                 
                if(graph[i][j] and graph[j][k] 
                   and graph[k][i]):
                    count = count+1
    count=count/3
    return count 
    
 

graph = [[0, 0, 1, 0],
        [1, 0, 0, 1],
        [0, 1, 0, 0],
        [0, 0, 1, 0]]
 
c = Triangles(graph)
c=int(c) 
print("Triangles present : ",c)
```

```python

class Node:
    count = 0
    def __init__(self, data):

        self.left = None
        self.right = None
        self.data = data
        

    def insert(self, data):
# Compare the new value with the parent node
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

# Print the tree
    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print( self.data),
        if self.right:
            self.right.PrintTree()

            
# Inorder traversal
# Left -> Root -> Right
    def inorderTraversal(self, root):
        res = []
        if root:
            res = self.inorderTraversal(root.left)
            res.append(root.data)
            res = res + self.inorderTraversal(root.right)
        return res        

# Preorder traversal
# Root -> Left ->Right
    def PreorderTraversal(self, root):
        res = []
        if root:
            res.append(root.data)
            res = res + self.PreorderTraversal(root.left)
            res = res + self.PreorderTraversal(root.right)
        return res   

# Postorder traversal
# Left ->Right -> Root
    def PostorderTraversal(self, root):
        res = []
        if root:
            res = self.PostorderTraversal(root.left)
            res = res + self.PostorderTraversal(root.right)
            res.append(root.data)
        return res

# findval method to compare the value with nodes
    def findval(self, lkpval):
        if lkpval < self.data:
            if self.left is None:
                return str(lkpval)+" is not Found"
            return self.left.findval(lkpval)
        elif lkpval > self.data:
            if self.right is None:
                return str(lkpval)+" is not Found"
            return self.right.findval(lkpval)
        else:
            return str(self.data) + " is found"
    
    def countNodes(self,root):
        count=1
        if root is None:
            return 0
        else:
            count += self.countNodes(root.left) 
            count += self.countNodes(root.right)
            return count
        
        
    def MirrorTree(self,root):
        if root is not None:
            if(root.left is not None and root.right is not None):
                print("Before Swapping\n")
                print(" Left : ",root.left.data)
                print(" Right : ",root.right.data)
                temp = root.left
                root.left = root.right
                root.right = temp
                print("After Swapping\n")
                print(" Left : ",root.left.data)
                print(" Right : ",root.right.data)
                self.MirrorTree(root.left)
                self.MirrorTree(root.right)
                return root
        else:
            return root


        
        
# Use the insert method to add nodes
root = Node(27)
root.insert(14)
root.insert(35)
root.insert(31)
root.insert(10)
root.insert(9)
root.insert(42)
print("Print the tree")
root.PrintTree()
print("Count of Nodes: ",root.countNodes(root))
root = root.MirrorTree(root)
root.PrintTree()
```

```python
#Pt1
from matplotlib import pyplot as plt
import pandas as pd

data=pd.read_csv("company_sales_data.csv")
#print(pd.DataFrame(data))
Months=data['month_number']
TotalProfit = data['total_profit']
#print("Months: ",Months)
#print("Total Profit",TotalProfit)
plt.plot(TotalProfit,Months,animated=True)
plt.title('Part 1')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
cream=data['facecream']
wash=data['facewash']
paste=data['toothpaste']
soap=data['bathingsoap']
sham=data['shampoo']
moist=data['moisturizer']
plt.plot(Months,cream,label='Face Cream', linewidth=6)
plt.plot(Months,wash,label='Face Wash', linewidth=6)
plt.plot(Months,paste,label='Toothpaste', linewidth=6)
plt.plot(Months,soap,label='Bath Soap', linewidth=6)
plt.plot(Months,sham,label='Shampoo', linewidth=6)
plt.plot(Months,moist,label='Moisturizer', linewidth=5)
plt.ylabel('Products')
plt.xlabel('Months')
plt.title('Part 2')
plt.legend()
plt.grid(True,color='k')
plt.show()
```
