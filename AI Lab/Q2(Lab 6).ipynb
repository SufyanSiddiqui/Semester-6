```python
import queue
from queue import Queue, PriorityQueue

def add_vertex(v,h):
  global graph
  global nod
  global vertices
  global heuristic_values
  if v not in vertices:
      nod = nod + 1
      vertices.append(v)
      heuristic_values.append(h)
      if nod > 1:
          for vertex in graph:
              vertex.append(0)
      temp = []
      for i in range(nod):
          temp.append(0)
      graph.append(temp)    
  else:
      print("Vertex already added") 

def add_edge(vertex1, vertex2, e):
    if vertex1 not in vertices:
        print("Vertex ", vertex1, " does not exist.")
    elif vertex2 not in vertices:
        print("Vertex ", vertex2, " does not exist.")
    else:
        index1 = vertices.index(vertex1)
        index2 = vertices.index(vertex2)
        graph[index1][index2] = e
        graph[index2][index1] = e

def Astar_search(start, goal):
    visited =[]
    path = []
    #openlist = []
    cost_gdbf = 0
    current_state = start
    path.append(current_state)
    visited.append(current_state)
    print("Current_state: ", current_state)
    while current_state != goal:
        s_index = 0
        openlist = []
        for s, i in enumerate(vertices):
            if i == current_state:
                s_index = s
                break
        for j in range(nod):
            print(s_index)
            if graph[s_index][j] != 0 and vertices[j] not in visited:
                print(vertices[j], '\t', heuristic_values[j])
                if vertices[j] != goal:
                    openlist.append(heuristic_values[j])
                    visited.append(vertices[j])
                elif vertices[j] == goal:
                    current_state = vertices[j]
                    path.append(current_state)
                    break
        if current_state == goal:
            minindex = heuristic_values.index(min(openlist))
            print("Minindex: ",minindex)
            current_state = vertices[minindex]
            if heuristic_values[minindex]!= 0: 
                cost_gdbf = cost_gdbf + heuristic_values[minindex]
                path.append(current_state)
                print("Current_state: ", current_state,"heuristic value:", heuristic_values[minindex])
            else:
                path.append(current_state)
                print("Minimum heuristic value reached")
                break
        else:
            break
    print("Path using A* Search method: ",path)
    print("Total cost: ",cost_gdbf)
        

vertices = []
heuristic_values = []
nod = 0
graph = []

nodes = ['S','A','B','C','D','G']
heu =   [5,3,4,2,6,0]


for h,n in enumerate(nodes):
  add_vertex(n,heu[h])

dict = {'S': [['A',1],['G',10]],
         'A': [['B',2], ['C',1]],
         'B': [['D',5]],
         'C': [['D',3],['G',4]],
         'D': [['G',4]]}

for k, v in dict.items():
    for i in range(len(v)):
        add_edge(k, v[i][0],v[i][1])

        
print(dict.items()) 
print("\nTotal nodes: ",nod)
print(graph)
print("\n",vertices)
print("\n", heuristic_values)
Astar_search('S','G')


```
