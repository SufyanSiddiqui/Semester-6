```python
import random as rn
import string as str
import numpy as np
import pandas as pd
Target = "ArtificialIntelligenceLab"
totalscore=0

def StrGen(size):
    chars=str.ascii_lowercase + str.ascii_uppercase #+ str.digits + str.ascii_letters
    return ''.join(rn.choice(chars) for _ in range(size))


def Fitness(population,Target):
    fitness_dict = dict()
    print(Target)
    
    totalscore=0
    for i in range(len(population)):
        score=0
      
        for j in range(len(population[i])):
            if Target[j] != population[i][j]:
                score+=1
            
        fitness_dict[population[i]]=score
        totalscore+=score
    return fitness_dict,totalscore

def Probability(fitness,total):
    ind=list()
    for key,value in fitness.items():
        ind.append(value)
    ind=np.array(ind)
    prob=ind/total
    #print("Probabilities")
    #print(prob)
    i=0
    for key,value in fitness.items():
        fitness[key]=[fitness[key],prob[i]]
        i+=1
    #print(fitness)
    return fitness
def Cross(p1,p2,start,end):
    c1=c2=''
    p1=list(p1)
    p2=list(p2)
    for i in range(start,end):
      p1[i],p2[i]=p2[i],p1[i]
      
    p1=''.join(p1)
    p2=''.join(p2)
    return p1,p2


def Crossover(fitness,start,end):
    lst=list(fitness.keys())
    n=0
    for i in range(len(lst)):
        if i+2>len(lst):
            return lst
        else:
            if lst[i]==lst[i+1]:
                n=i+2
            else:
                n=i+1
                

            p1=lst[i]
            p2=lst[i+1]
            p1,p2 = Cross(p1,p2,start,end)
            lst[i]=p1
            lst[i+1]=p2
            i=n

    #print(lst)

    return lst
def Mutation(lst):
    for i in range(len(lst)):
        lst2=list(lst[i])
        #print(lst2)
        n=rn.randint(0,22)
        c=rn.choice(str.ascii_letters)
        lst2[n]=c
        lst[i]=''.join(lst2)
    #print(lst)
    return lst


    


        








population=list()
for i in range(70):
    s=StrGen(25)
    population.append(s)

for i in range(10):
    popul_d,totalscore=Fitness(population,Target)
#print(popul_d)
#totalscore
    popul_d=Probability(popul_d,totalscore)
    popul_d = dict(sorted(popul_d.items(),key=lambda item: item[1]))
    print(popul_d)
#print(Target)
    population=Crossover(popul_d,5,10)
    population=Mutation(population)
#print(new_d)
    #lst=list(popul_d.items())
    #print(lst)
    #print(population)





#len(population)
```
